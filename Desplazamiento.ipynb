```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from google.colab import files # Solo para Colab

# 1. Subir video
print("Sube tu video de la lavadora (1920x1200 @ 90fps):")
uploaded = files.upload()
video_name = list(uploaded.keys())[0]
print(f"Archivo {video_name} cargado con éxito.")

# 2. Extraer primer frame para calibrar
cap = cv2.VideoCapture(video_name)
fps_camera = 90  # Basado en tus specs
success, first_frame = cap.read()
if not success:
    print("Error al leer el video.")
else:
    # Convertir a RGB para mostrarlo correctamente en las gráficas
    first_frame_rgb = cv2.cvtColor(first_frame, cv2.COLOR_BGR2RGB)
    plt.imshow(first_frame_rgb)
    plt.title("Primer Frame del Video")
    plt.show()
```
